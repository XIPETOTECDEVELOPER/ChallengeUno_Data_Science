### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# VENTAS TIENDA 1
# Llamo a la columna precio y sumo todas las ventas
# Como resultado tengo el total de ventas

suma_T1 = sum(tienda["Precio"])
print(f'El total de ventas de la Tienda 1 es: ${suma_T1}')
```

```python
# VENTAS TIENDA 2
# Llamo a la columna precio y sumo todas las ventas
# Como resultado tengo el total de ventas

suma_T2 = sum(tienda2["Precio"])
print(f'El total de ventas de la Tienda 2 es: ${suma_T2}')
```

```python
# VENTAS TIENDA 3
# Llamo a la columna precio y sumo todas las ventas
# Como resultado tengo el total de ventas

suma_T3 = sum(tienda3["Precio"])
print(f'El total de ventas de la Tienda 3 es: ${suma_T3}')
```

```python
# VENTAS TIENDA 4
# Llamo a la columna precio y sumo todas las ventas
# Como resultado tengo el total de ventas

suma_T4 = sum(tienda4["Precio"])
print(f'El total de ventas de la Tienda 4 es: ${suma_T4}')
```

```python
# Creo un diccionario de las tiendas con sus
# ventas correspondientes a cada tienda.

sumaTotal = {'Tienda 1' : suma_T1, 'Tienda 2' : suma_T2, 'Tienda 3' : suma_T3, 'Tienda 4' : suma_T4}
sumaTotal
```

```python
# Creo una funcion para sumar todas las ventas
# Como resultado obtengo el total de las ventas de todas las tiendas

sumaTT = 0
for s in sumaTotal.values():
  sumaTT += s

print(f'El total de ventas por las 4 tiendas son: ${sumaTT}')
```

```python
# De nuestro diccionario vamos a obtener el min para saber que tienda es la que menos vende

print(min(sumaTotal.values()))
```

```python
if sumaTotal.items() == 1038375700.0:
  print("La tienda con menores ventas es: Tienda 4")
```

```python
# De nuestro diccionario vamos a obtener el max
# para saber que tienda es la que vende más

print(max(sumaTotal.values()))
```

```python
# Grafica Ventas
plt.figure(figsize=(10, 6))

x = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'] # Tiendas
y = [1150880400, 1116343500.0, 1098019600, 1038375700] # Ventas (data numérica)

# agregar nombre a los ejes y titulo al grafico
plt.xlabel('Tiendas')
plt.ylabel('Ventas')
plt.title('Ventas Por Tienda')
plt.bar(x, y)
plt.show()
```

```python
# Como resultado y visualizando la gráfica obtenemos que la tienda con menor ventas
# Es la Tienda 4
```

# 5. Envío promedio por tienda

##### *Tienda 1*

```python
# Vamos a Sacar el promedio de "Costos de Envio por Tienda"
# Mando a llama a la columna "COSTOS DE ENVIO" de la Tienda 1
# Con la función (Sum), vamos a sumar todos los envio de la Columna
# E imprimimos el resultado de la Suma con la funcion (print)
# Como resultado obtenemos lo que le cuesta a la Tienda 1 los Costos de Envio

sumaPromedio_enviosTienda1 = sum(tienda["Costo de envío"])
print(f'El total de "COSTOS DE ENVIO" de la Tienda 1 es: ${sumaPromedio_enviosTienda1}')
```

```python
# Como ya sabemos cuanto es lo que le cuesta a la Tienda 1 Enviar los productos
# Necesitamos sacar el promedio de estos costos, ya que nos los pide el Señor Juan Perez Lopez de Colombia
# Como resultado tenemos el promedio de envios de la Tienda 1
# Con esto haremos una comparacion de promedios de COSTOS DE ENVIO con las demás tiendas.

promedioT1_CE = sum(tienda['Costo de envío']) / len(tienda['Costo de envío'])
print(f'El promedio de "COSTOS DE ENVIO" de la tienda 1 es: ${round(promedioT1_CE)}')
```
